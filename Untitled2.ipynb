## Najbolj priljubljene številke

V tem poglavju sem želel ugotoviti, katere številke igralci najraje izbirajo za "svojo". Prav tako me je zanimalo na podlagi česa se odločajo za dano številko. Moje domneva je namreč, da se igralci za določeno številko odločijo tudi na podlagi svojega rojstnega datuma. 

```python
igralci_po_stevilkah = stevilke.groupby('stevilka').size()
igralci_po_stevilkah.sort_values(ascending=False).head(25).plot.bar(x='ime', y='stevilo_igralcev')
plt.title('Največkrat nošene številke', fontsize=18)
plt.xlabel('številka', fontsize=14)
plt.ylabel('število sezone nošenja številke', fontsize=14)
```

Najbolj nošena število od leta 1964 naprej v ligi NBA je število 5. To pomeni, da je dano število bilo nošeno v največ sezonah, skupno v kar 699-tih. Tu naprimer, če imata dva igralca iz različnih ekip isto sezono nosita isto število, to šteje za dve sezoni. Zanimivo je, da so prva 4 najbolj nošena števila liha.

```python
stevilke_razlicni_igralci = stevilke.groupby('stevilka')['ime'].nunique()
stevilke_razlicni_igralci.sort_values(ascending=False).head(25).plot(kind="bar")
plt.title('Največkrat izbrane številke', fontsize=18)
plt.xlabel('številka', fontsize=14)
plt.ylabel('število igralcev', fontsize=14)
```

Število, katerega je nosilo največ igralcev je število 11. Razlika med številom 5 in 11 je torej v tem, da je določen igralec število 5 nosil dlje kot število 11, torej so bili igralci tej številki v nekem smislu "bolj zvesti". Opaziti se da tudi to, da je 5 najbolj nošenih števil izbralo tudi največ različnih igralcev.

```python
stevilke_razlicni_igralci.sort_values(ascending=True).count()
neuporabljene = []
for stevilo in range(0, 100):
    if stevilo not in np.unique(stevilke['stevilka']):
        neuporabljene.append(stevilo)
neuporabljene
```

## Najuspešnejše številke

Pri analizi uspešnosti posameznih številk sem se osredotočil le na številke, katere je nosilo več kot 10 igralcev, sej se mi sicer podatki niso zdeli legitimni. Izstopale so namreč številke, katere je nosil le eden igralec, na primer številka 84 katero je v sezoni 2006-2007 nosil igralec Chris Webber.

```python
dff = pd.DataFrame([stevilke.groupby('stevilka')['ime'].nunique()]).T
zdruzen = zdruzen.join(dff, on='stevilka', how='left', lsuffix="_review")
```

```python
zdruzen.columns = ['ime', 'konec_sezone', 'ekipa', 'stevilka', 'visina', 'teza', 'leto_rojstva', 'mesec_rojstva', 'dan_rojstva', 'povprecje_tocke', 'povprecje_skoki', 'povprecje_asistence', 'stevilo_igralcev_s_stevilko']
zdruzen_analiza = zdruzen[zdruzen.stevilo_igralcev_s_stevilko > 10]
```

### Številka dresa v povezavi s točkami

```python
zdruzen_analiza.groupby('stevilka')['povprecje_tocke'].mean().sort_values(ascending=False).head(20).plot(kind="bar")
plt.title('Številke z največjim povprečjem točk na tekmo', fontsize=18)
plt.xlabel('številka dresa', fontsize=12)
plt.ylabel('povprečno število doseženih točk', fontsize=12)
```

Najuspešnejši strelci na koš so v povprečju igralci, ki nosijo številko 16. Številke so glede na povprečje točk na tekmo precej raznolike, saj najmanjšo številko predstavlja 0, največjo pa 42. 

### Številka dresa v povezavi z asistencami

```python
zdruzen_analiza.groupby('stevilka')['povprecje_asistence'].mean().sort_values(ascending=False).head(20).plot(kind="bar")
plt.title('Številke z največjim povprečjem asistenc na tekmo', fontsize=18)
plt.xlabel('številka dresa', fontsize=12)
plt.ylabel('povprečno število doseženih asistenc', fontsize=12)
```

Pri analizi številke dresa v povezavi s povprečnim številom asistenc na tekmo lahko opazimo, da so številke v povprečju precej nižje kot pri točkah. Najnižja številka je tu 1, najvišja pa le 24. Pri doseganju asistenc so običajno uspešnejši organizatorji igre (angleško imenovani "playmaker"-ji). Z izjemo nekaterih izjem, so ti igralci navadno nižji in posledično precej bolj spretni. Med izjeme spada na primer Magic Johnson, ki z 208 centimetri velja za enega najboljših podajalcev vesh časov. Danes igra v najmočnejši košarkarski ligi na svetu tudi Srb Nikola Jokić, ki med centri velja za enega boljših asistentov. V tej sezoni kljub 208 centimetrom in 113 kilogramom dosega v povprečju kar 5.5 podaje na tekmo.

```python
zdruzen[zdruzen.visina >= 200].groupby('ime')['visina', 'povprecje_asistence', 'teza'].mean().sort_values(ascending=False, by = 'povprecje_asistence').head(5)
```

Svoje nadaljno delo sem tako usmeril v raziskovanje karakteristik dobrega organizatorja igre. Na spodnje grafu sem tako predstavil povprečno število asistenc doseženih na tekmo v odvisnosti od številke igralca, kjer sem točke pobarval glede na višino igralca.

```python
zd1 = zdruzen_analiza.groupby('stevilka')['povprecje_tocke', 'povprecje_asistence', 'povprecje_skoki', 'visina'].mean().sort_values(ascending=False, by = 'povprecje_asistence')
zd1.reset_index(level=0, inplace=True)
```

```python
plt.scatter(zd1.stevilka, zd1.povprecje_asistence, c=zd1.visina, s=50)
plt.title('Številke v povezavi z asistencami', fontsize=18)
plt.xlabel('številka', fontsize=14)
plt.ylabel('povprečje asistenc na tekmo', fontsize=14)
plt.colorbar()
```

V skladu s prejšnjimi opažanji nižji igralci v povprečju res izbirajo nižje številke in v povprečju dosegajo tudi več asistenc, medtem ko števile nad 30 izbirajo večinoma le višji igralci. V nadaljevanju sem analiziral korelacijo med karakteristikami igralca in njegovimi asistencami.  

```python
zdruzen.plot.scatter('teza', 'povprecje_asistence')
plt.title('Povezava med težo in asistencami', fontsize=18)
plt.xlabel('telesna teža', fontsize=12)
plt.ylabel('povprečno število doseženih asistenc', fontsize=12)
zdruzen.plot.scatter('visina', 'povprecje_asistence')
plt.title('Povezava med višino in asistencami', fontsize=18)
plt.xlabel('telesna višina', fontsize=12)
plt.ylabel('povprečno število doseženih asistenc', fontsize=12)
```

Opazimo, da število dosečenih asistenc tako s telesno višino kot tudi s telesno težo upada. Seveda so prisotne tudi izjeme, katere sem omenil že prej. 

### Številka dresa v povezavi s skoki

```python
zdruzen_analiza.groupby('stevilka')['povprecje_skoki'].mean().sort_values(ascending=False).head(20).plot(kind="bar")
plt.title('Številke z največjim povprečjem skokov na tekmo', fontsize=18)
plt.xlabel('številka dresa', fontsize=12)
plt.ylabel('povprečno število doseženih skokov', fontsize=12)
```

Največ skokov v skladu s pričakovanji dosegajo višji igralci, ki v povprečju izbirajo tudi višje številke. To je dobro razvidno tudi iz zgornjega grafa.

Povprečno število skokov glede na telesno višino igralca narašča skladno s pričakovanji. Zanimivo pa je, da sta največ skokov na tekmo v svoji karieri dosegala igralca nižja od 220 cm.
